```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
data = np.load('chestmnist.npz')
image_title = 'Satvik_Nayak_'

# print(data.files) -> ['train_images', 'val_images', 'test_images', 'train_labels', 'val_labels', 'test_labels']

train_images = data['train_images']
img = train_images[0, :, :]

# print('shape:', img.shape)

plt.imshow(img, cmap='gray')
plt.savefig(image_title + '1.png')
plt.show()

plt.imshow(img, cmap='binary')
plt.savefig(image_title + '2.png')
plt.show()

img_transform = cv2.boxFilter(img, -1, (3, 3)) # blurs image
img_transform = cv2.flip(img_transform, 1) # horizontal flip
plt.imshow(img_transform, cmap='gray')
plt.savefig(image_title + '3.png')
plt.show()

# To display more images
# for int in range(0, 100):
#     print(int)
#     tmp = train_images[int, :, :]
#     plt.imshow(tmp, cmap='gray')
#     plt.show()
```
